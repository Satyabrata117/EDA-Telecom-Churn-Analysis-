# **Conclusion**

In the course of conducting exploratory data analysis (EDA) on our dataset, several key insights have emerged, shedding light on factors influencing customer churn. The following is a comprehensive summary of our findings:

1. Regional Disparities in Churn Rates:
Certain states exhibit significantly higher churn rates, suggesting a potential correlation with suboptimal network coverage in those areas.

2. Identification of Redundant Data Columns:
Through our analysis, we have identified redundant features in the dataset. Specifically, both area code and account length were found to lack a discernible impact on the churn rate, rendering them redundant for predictive modeling.

3. Impact of International Plan on Churn Rates:
Customers enrolled in the International plan showcase a notably elevated churn rate. This phenomenon points towards dissatisfaction, potentially stemming from network issues and the imposition of high international calling charges.

4. Voice Mail Quality as a Churn Indicator:
A noteworthy observation is the relationship between churn and the number of voice-mail messages exceeding 20. This suggests a potential dissatisfaction with the quality of voice mail services offered.

5. Irrelevance of Call-related Columns:
Several columns pertaining to call metrics during different times of the day (Total day, eve, night) were found to be irrelevant in predicting churn. Streamlining the dataset by excluding these variables could enhance the predictive accuracy of our model.

6. High Churn in International Calls:
Notably, customers with an international plan experience a heightened churn rate. This underscores concerns regarding the cost-effectiveness of international calls or potential network issues during such calls.

7. Impact of Customer Service Calls:
An intriguing finding is the correlation between customer churn and calls to the service center. Elevated churn rates following customer service calls indicate a potential gap in issue resolution, emphasizing the need for enhanced customer support.

In conclusion, our exploratory data analysis has unearthed actionable insights that can inform strategic decisions aimed at mitigating churn. Addressing issues related to network coverage, international calling charges, voice mail quality, and optimizing customer service responses can collectively contribute to a more robust and customer-centric service model. Furthermore, the identification of redundant features allows for a streamlined dataset, refining the foundation for our predictive modeling efforts. These insights provide a solid groundwork for further investigation and targeted interventions to improve overall customer satisfaction and retention.
