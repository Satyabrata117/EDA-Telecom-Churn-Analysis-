## Graphical Analysis

UNIVARIATE ANALYSIS

In Univariate Analysis we analyze data over a single column from the numerical dataset, for this we use 3 types of plot which are box plot, strip plot, dis plot.

````python
#Printing boxplot for each numerical column present in the data set
df1=telecom_df.select_dtypes(exclude=['object','bool'])
for column in df1:
        plt.figure(figsize=(17,1))
        sns.boxplot(data=df1, x=column)
plt.show()
````

````python
#Printing displot for each numerical column present in the data set
df1 = telecom_df.select_dtypes(exclude=['object', 'bool'])

# Calculate the number of rows and columns for subplots
num_rows = len(df1.columns) // 2 + len(df1.columns) % 2
num_cols = 2

# Set up the figure and axes
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 3 * num_rows))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop through numerical columns and create histograms
for i, column in enumerate(df1.columns):
    sns.histplot(data=df1, x=column, ax=axes[i], kde=True)
    axes[i].set_title(column)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
````

````python
#Printing strip plot for each numerical column present in the data set
df1=telecom_df.select_dtypes(exclude=['object','bool'])
for column in df1:
        plt.figure(figsize=(17,1))
        sns.stripplot(data=df1, x=column)
plt.show()
````


## BIVARIATE ANALYSIS

In Bivariate Analysis we analyze data by taking two columns into consideration from a dataset, here we only take numerical data type column, for this visualization we use Box plot,scatter plot

````python
# Plot a boxplot for churn column  by each numerical feature present in the data set
df2= telecom_df.describe().columns
for col in df2:
  fig=plt.figure(figsize=(17,3))
  ax=fig.gca()
  feature=telecom_df[col]
  label=telecom_df['Churn']
  correlation= feature.corr(label)
  plt.scatter(x=feature,y=label)
  plt.xlabel(col)
  plt.ylabel('Churn')
plt.show()
````

## Multivariate Analysis

In Multivariate Analysis we analyze data by taking more than two columns into consideration from a dataset,for this we using correlation plot,correlation matrix, correletaion heatmap, pair plot

````python
plt.figure(figsize=(19,8))
telecom_df.corr()['Churn'].sort_values(ascending = False).plot(kind='bar',color = ['red','blue','yellow','indigo','orange','brown','pink'])
````

````python
# Set the figure size
plt.figure(figsize=(19, 8))

# Calculate the correlation with Churn
correlation_series = telecom_df.corr()['Churn'].sort_values(ascending=False).drop('Churn')

# Create a bar plot using Seaborn
sns.barplot(x=correlation_series.values, y=correlation_series.index, palette="coolwarm")

# Set labels and title
plt.xlabel('Correlation with Churn')
plt.ylabel('Features')
plt.title('Correlation Plot with Churn')

# Show the plot
plt.show()
````

````python
## plot the Correlation matrix
plt.figure(figsize=(17, 8))

# Calculate the correlation matrix using numeric_only=True
correlation = telecom_df.corr(numeric_only=True)

# Plot the heatmap
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Set the title
plt.title('Correlation Matrix')

# Show the plot
plt.show()
````

````python
#create a correlation heatmap
#Assigning true=1 and False=0 to churn variable
telecom_df['Churn'] = telecom_df['Churn'].replace({bool(True):1,bool(False):0})
plt.figure(figsize=(17,9))
sns.heatmap(telecom_df.corr(), cmap="Paired")
plt.title("Correlation Heatmap", fontsize=20)
````

# **Conclusion**

The exploratory data analysis (EDA) conducted on our telecom dataset has revealed pivotal insights, shedding light on critical factors influencing customer churn. This summary encapsulates the key findings and proposed actions to fortify our customer retention strategies:

Regional Dynamics:

Insight: Churn rates vary across states, indicating potential issues with network coverage.
Action: Implement targeted improvements in states with high churn rates to enhance network infrastructure and overall customer satisfaction.
Redundant Features:

Insight: Area code and account length are non-contributing factors to churn.
Action: Streamline model development by excluding these redundant features, optimizing predictive accuracy without unnecessary complexity.
International Plans and Charges:

Insight: Customers with international plans exhibit higher churn rates, possibly due to dissatisfaction with network issues and international call charges.
Action: Reassess international plans and address network quality to align offerings with customer expectations and reduce churn.
Voice Mail Quality:

Insight: Churn increases when voice mail messages exceed 20, highlighting potential dissatisfaction with voice mail quality.
Action: Enhance voice mail services to improve quality, ensuring a positive customer experience and mitigating unnecessary churn.
Non-Influential Metrics:

Insight: Total day and night metrics have negligible impact on churn prediction.
Action: Deprioritize these metrics in future analyses to focus on more influential factors affecting customer attrition.
International Call Challenges:

Insight: Higher churn rates among international plan customers indicate potential issues with call drops or network problems.
Action: Conduct a thorough review of international call service quality, addressing challenges to enhance customer satisfaction and retention.
Customer Service Calls:

Insight: Increased churn when unsatisfied customers contact the service center underscores the need for more efficient issue resolution.
Action: Optimize customer service processes to promptly resolve issues, preventing churn and fostering positive customer relationships.
In conclusion, these insights form the basis for strategic initiatives aimed at reducing churn and fortifying customer loyalty. By implementing targeted improvements in network coverage, service plans, and customer service, we position ourselves to meet and exceed customer expectations, ensuring the long-term success of our telecom services.
